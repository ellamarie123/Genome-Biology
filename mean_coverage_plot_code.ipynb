```python
# Import pandas module 
import pandas as pd

# Load the filtered data
df = pd.read_csv('filtered_coverage_report.txt', sep='\t')

# Display the first few rows of the dataframe
df.head()
```

```python
# Load the unfiltered data 
df_two = pd.read_csv('coverage_report.txt', sep='\t')

# Display the first few rows of the second dataframe
df_two.head()
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Plot histogram for unfiltered dataset
plt.hist(data=df_two, x='coverage', bins=10, alpha=0.5, label='Before Filtering')

# Plot histogram for filtered dataset
plt.hist(data=df, x='coverage', bins=10, alpha=1, label='After Filtering')

# Add vertical lines for mean coverage
plt.axvline(x=73.52, color='blue', linestyle='--', label='Mean Before Filtering',  ymin=0, ymax=0.75)
plt.axvline(x=99.78, color='orange', linestyle='--', label='Mean After Filtering', ymin=0, ymax=0.75)

# Add labels and legend
plt.xlabel('Coverage')
plt.ylabel('Frequency')
plt.title('Distribution of Coverage Before and After Filtering')
plt.legend()

# Show plot
plt.show()
```
